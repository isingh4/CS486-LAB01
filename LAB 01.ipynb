### HEATMAP

```python
plt.figure(figsize=(40, 24))
sns.heatmap(corr, annot=True, annot_kws={'size':13})
plt.title("H E A T M A P  O F  C O R R E L A T I O N S  B E T W E E N  V A R I A B L E S", fontsize =40)
```
